```python
item = '19M'
item[-1]
```

```python
def size_process(item):
  if str(item)[-1] == 'M':
    res = float(str(item).replace('M',''))
    res = res*1024
    return res
  elif str(item)[-1] == 'K':
    res = str(item).replace('K','')
    return res
  else:
    return str(np.NaN)
```

```python
GPS.Size = GPS.Size.apply(size_process)
```

```python
GPS.Size = GPS.Size.astype('float')
```

```python
item = '$14.00'
str(item)[0]
```

```python
def price_process(item):
  if str(item)[0] == '$':
    wto = str(item).replace('$','')
    return wto
  else:
    return item
```

```python
GPS.Price = GPS.Price.apply(price_process)
```
